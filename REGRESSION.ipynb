1.  What is Simple Linear Regression?
- Simple Linear Regression is a statistical method used to model the relationship between two variables:

One independent variable (X) ‚Üí the predictor

One dependent variable (Y) ‚Üí the outcome

It helps us understand how changes in X affect Y and allows us to make predictions.

2.  What are the key assumptions of Simple Linear Regression ?
- Key Assumptions of Simple Linear Regression

For Simple Linear Regression to give reliable results, these main assumptions must be satisfied:

1Ô∏è‚É£ Linearity

There must be a linear relationship between the independent variable (X) and the dependent variable (Y).

The data points should roughly form a straight-line pattern.

If the relationship is curved, linear regression is not appropriate.
  -->

2Ô∏è‚É£ Independence of Errors

The residuals (errors) should be independent of each other.

One observation should not influence another.

Important in time-series data (where autocorrelation may occur).

3Ô∏è‚É£ Homoscedasticity (Constant Variance)

The variance of residuals should be constant across all values of X.

The spread of residuals should look roughly the same everywhere.

If the spread increases or decreases, it‚Äôs called heteroscedasticity.

4
. Normality of Errors

The residuals should be approximately normally distributed.

This is especially important for hypothesis testing and confince intervals.

Can be checked using a histogram or Q-Q plot.

5Ô∏è‚É£ No Perfect Multicollinearity

In simple linear regression, there is only one predictor, so this assumption is automatically satisfied.
(In multiple regression, predictors should not be highly correlated.)

9. What is the main difference between Simple and Multiple Linear Regression?
  - Simple Linear Regression and Multiple Linear Regression are both methods used to study the relationship between variables and to make predictions. However, they differ mainly in the number of independent variables they use and in how they explain the outcome.

Simple Linear Regression

Simple Linear Regression uses only one independent variable to predict a dependent variable. It examines how a single factor influences an outcome. The relationship between the two variables is represented by a straight line.

For example, if we want to predict a student‚Äôs exam score based only on the number of hours studied, we are using simple linear regression. In this case, we assume that exam performance depends only on study hours.

Graphically, this relationship is shown as a straight line on a two-dimensional graph.

4

This method is easier to compute and interpret because there is only one predictor involved.

Multiple Linear Regression

Multiple Linear Regression uses two or more independent variables to predict a dependent variable. It helps us understand how several factors together influence the outcome.

For example, if we want to predict a student‚Äôs exam score based on:

Hours studied

Attendance

Practice test performance

then we use multiple linear regression. This method allows us to see the effect of each factor while keeping the others constant.

When there are two predictors, the relationship forms a plane in three-dimensional space. With more predictors, the model extends into higher dimensions.

10.  What are the key assumptions of Multiple Linear Regression?
- Multiple Linear Regression (MLR) is used to study the relationship between one dependent variable and two or more independent variables. For the model to produce reliable and valid results, certain assumptions must be satisfied.

1. Linearity

The relationship between the dependent variable and each independent variable should be linear.

This means that changes in the predictors should produce proportional changes in the outcome variable. If the relationship is curved or non-linear, the linear regression model may not fit the data properly.

2. Independence of Errors

The residuals (errors) should be independent of each other.

This means that the error for one observation should not be related to the error for another observation. This assumption is especially important in time-series data, where consecutive observations might be correlated.

3. Homoscedasticity (Constant Variance of Errors)

The variance of the residuals should remain constant across all levels of the independent variables.

In simple terms, the spread of the errors should be roughly the same for all predicted values. If the spread increases or decreases (forming a funnel shape), the assumption is violated. This problem is called heteroscedasticity.
11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?
- Heteroscedasticity occurs when the variance of the residuals (errors) is not constant across all levels of the independent variables.

In simpler terms:

The spread of the errors changes depending on the value of the predictors.

Homoscedasticity ‚Üí constant variance of errors

Heteroscedasticity ‚Üí unequal or varying variance of errors
How Heteroscedasticity Affects Regression Results

Invalid Standard Errors

Standard errors of the regression coefficients may be biased.

This affects t-tests and confidence intervals, making them unreliable.

Incorrect Significance Tests

Coefficients might appear statistically significant when they are not (Type I error), or vice versa.

Reduced Predictive Accuracy

Predictions may be less reliable, especially for values where residual variance is large.

Model Assumptions Violated

One of the key assumptions of regression is homoscedasticity.

Violating it can weaken the interpretability of the model.

‚úÖ Ways to Detect Heteroscedasticity

Residuals vs Fitted Values Plot ‚Üí look for a funnel shape

Breusch-Pagan Test ‚Üí statistical test for non-constant variance

White Test ‚Üí another formal test for heteroscedasticity

üîß How to Fix or Reduce Heteroscedasticity

Transform the Dependent Variable

Use log, square root, or Box-Cox transformation.

Weighted Least Squares Regression

Give less weight to observations with larger variance.

Add or Adjust Predictors

Sometimes missing variables cause changing variance.

Robust Standard Errors

Adjust standard errors to account for heteroscedasticity without changing coefficients.

12.  How can you improve a Multiple Linear Regression model with high multicollinearity?
- Multicollinearity occurs when two or more independent variables in a Multiple Linear Regression model are highly correlated, making it hard to estimate their individual effects accurately. High multicollinearity can inflate standard errors and make coefficients unstable.

Here‚Äôs how to address it:

1Ô∏è‚É£ Remove Highly Correlated Predictors

Identify predictors that are strongly correlated using a correlation matrix or Variance Inflation Factor (VIF).

Remove one of the correlated variables to reduce redundancy.

Example: If both weight and BMI are in the model, you might keep only one.

2Ô∏è‚É£ Combine Variables

If variables measure similar concepts, combine them into a single composite variable.

Example: Combine height and weight into body mass index (BMI).

3Ô∏è‚É£ Use Principal Component Analysis (PCA)

PCA transforms correlated predictors into a smaller set of uncorrelated components.

Regression is then performed on these components.

Useful when there are many predictors and complex correlations.

4Ô∏è‚É£ Regularization Techniques

Ridge Regression: Adds a penalty term to shrink coefficients of correlated predictors.

Lasso Regression: Can shrink some coefficients to zero, effectively performing variable selection.

Both reduce variance caused by multicollinearity and improve prediction accuracy.

5Ô∏è‚É£ Center or Standardize Variables

Subtract the mean (center) or scale (standardize) the predictors.

This doesn‚Äôt remove multicollinearity but can reduce numerical instability in calculations.

6Ô∏è‚É£ Collect More Data

Increasing sample size can sometimes reduce multicollinearity‚Äôs effect and stabilize coefficient estimates.

üîé How to Detect Multicollinearity

Correlation matrix: Look for high correlations (> 0.8 or 0.9) between predictors.

Variance Inflation Factor (VIF):

VIF > 5 ‚Üí moderate multicollinearity

VIF > 10 ‚Üí high multicollinearity

13.  What are some common techniques for transforming categorical variables for use in regression models?
-  Common Techniques for Transforming Categorical Variables in Regression Models

Regression models, like linear regression, require numerical input. Categorical variables (like gender, color, or region) need to be transformed into numbers before they can be used. Here are the most common techniques:

1Ô∏è‚É£ One-Hot Encoding (Dummy Variables)

Converts each category into a binary variable (0 or 1).

2Ô∏è‚É£ Label Encoding

Assigns numerical values to categories.

Example: Size {Small, Medium, Large} ‚Üí {1, 2, 3}

Caution: This implies an order or magnitude, so it‚Äôs suitable only for ordinal variables (like Low, Medium, High) and not nominal ones.

3Ô∏è‚É£ Ordinal Encoding

Similar to label encoding but explicitly respects the order of categories.

Often used for variables like:

Education level (High School < Bachelor < Master < PhD)

Rating scales (Poor < Average < Good < Excellent)

4Ô∏è‚É£ Frequency Encoding

Replace each category with its frequency or proportion in the dataset.

Example: If ‚ÄúRed‚Äù appears 30% of the time, ‚ÄúBlue‚Äù 50%, ‚ÄúGreen‚Äù 20%, then values become {0.3, 0.5, 0.2}.

Useful when there are many categories.

5Ô∏è‚É£ Target Encoding (Mean Encoding)

Replace each category with the mean of the target variable for that category.

Example: If average house price in ‚ÄúDowntown‚Äù = 300k, ‚ÄúSuburb‚Äù = 200k, then categories become {300000, 200000}.

Caution: Can cause data leakage; apply carefully using cross-validation.

6Ô∏è‚É£ Binary Encoding

Converts categories into binary code using fewer columns than one-hot encoding.

Efficient for high-cardinality categorical variables (like ZIP codes or product IDs).

It measures how much the dependent variable (Y) changes for a one-unit increase in the independent variable (X), holding other variables constant (in multiple regression).

Direction of Relationship

Positive slope (b > 0): Y increases as X increases.

Negative slope (b < 0): Y decreases as X increases.
